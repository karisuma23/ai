```python
# # (영어) 알파벳과 숫자를 섞어서 입력을 받아서 숫자와 문자를 
# # 분리하여 출력하는 프로그램을 작성하시오.

# [출력] 

# 문자:
# 숫자:

```

```python
str = input('알파벳과 숫자를 입력하시오')

al = ''
no = ''
for i in str:
   if i.isalpha() :
     al = al+ i
   elif i.isdigit(): 
       no = no + i

print('문자: ', al)
print('숫자 :', no)
```

```python
def disting( ur ):
    al = ''
    no = ''

    for i in ur:
        if i.isdigit():
            al += i
            
        elif i.isalpha():
            no += i
                                    
    print(f'숫자 : {al} {len(al)}개 ');
    print(f'문자 : {no} {len(no)}개');
            
ur = input("입력 > ");
disting(ur);
```

```python
def disting(ur):
  al = ''
  no = ''

  for i in ur:
      if i in 'qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM':
          al += i
            
      elif i in '1234567890':
            no += i
                                    
  print(f'숫자 : {al} {len(al)}개 ');
  print(f'문자 : {no} {len(no)}개');
            
ur = input("입력 > ");
disting(ur);
  
```

```python
def ord_print( str ):
    al = ''
    no = ''

    for i in str :
      if 64< ord(i) <123:
       al=al+i
    
      elif 48< ord(i) <58:
        no=no+i
                                    
    print(f'숫자 : {al} {len(al)}개 ');
    print(f'문자 : {no} {len(no)}개');
            
str = input("입력 > ");
ord_print(str);   
```

```python
######################################
# 알고리즘 
######################################
# '알고리즘은 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표현한
# 아이디어나 생각'
# 알고리즘 표현 방법은 수도코드(pseudo code), 순서도(flow chart)이 있다.

# 프로그램 작성 순서
# 기회  - 설계(알고리즘) - 프로그래밍 - 디버그 - 문서작성  

# 좋은 알고리즘 
# 알기 쉽고, 속도가 빠르고 , 효율적이고(메모리를 적게사용), 재이용이 쉽다.

# 왜 알고리즘 ??
# 좋은 프로그램을 만들수 있고, 좋은 프로그램인지 판단할수 있고
# 프로그램 작성을 효율적으로 할수 있고 프로그래밍 기술을 향상시키는데 큰 도움이 된다.

# 좋은 알고리즘의 조건
# 결과를 얻을수 있어야 하고 반드시 종료되어야 한다.

# 알고리즘의 세가지 기본형
# 1 순차구조 : 처음부터 순서대로 처리
# 2 선택구조 : 조건식 if
# 3 반복구조 : 반복문 for while 


# 순서도 flow 차트, 흐름도 
# 프로그래밍 언어를 사용하지 않고 알고리즘을 기술하는 방법 
# 처리와 정차를 시각적으로 표현 한다.
# 자신이 생각하고 있는 알고리즘이 올바른지, 처리와 절차에 문제가 없는지를
# 눈으로 확인할수 있고 
# 주의 사항
# 처리 기호나 판단 기호가 많아지면 흐름이 복잡하게 꼬인다.
     # 1. 입력은 반드시 위에서 들어가도록 만든다
     # 2. 판단기호로 부터 나오는 흐름선에는 YES NO를 확실하게 기입한다.
     # 3. 입력선이나 출력선이 없는 처리 기호는 만들지 않는다. 
```

```python
# 이진 탐색법 Binary Search

# 탐색 대상 데이터들이 미리 오름차순 또는 내림차순으로 정렬이 되어 있어야만 한다. 
# 탐색하는 범위를 절반씩 추려나가면서 탐색을 한다.  

0  1   2   3   4  5   6  index 
11 13 17 19 23 29 31 data


0  1  2   3   4   5   6      step 1 가운데 인덱스  즉 3을 열여서 확인한다. 
           19                            찾고 있는 17이 아니다. 그런데 정렬되어 있다는 전제
                                          조건이 있으므로 이후의 모든 데이터는 절대 찾고 있는
                                          17이 될수 없다. 
0   1   2                       step 2 새로운 범위의 가운데 인덱스 즉 1을 열어 확인한다. 
    13                                    찾고 있는 17이 아니다. 그런데 정렬되어 있다는 전제
                                          조건이 있으므로 이전의 모든 데이터는 절대 찾고 있는
                                          17이 될수 없다. 
2                                step 3 마지막 요소만 남게 되었지만 찾고 있는 17이 아예 없는
17                                        경우도 발생할 수 있으므로 처리해 주어야 하지만 지금은
                                          마지막 index 2가 찾고 있는 17이므로 검색을 종료한다. 
                                                                                    
```

```python
# 이진 탐색법의 알고리즘
# 1. 가운데 요소를 선택하는 처리(인덱스의 평균을 계산하여 구한다. )
# 2. 가운데 데이터와 찾는 데이터를 비교하는 처리
# 3. 탐색 범위를 절반으로 줄이는 처리 

# 요소의 갯수가 짝수 일때는 어떻게 해야 할까!!!
            # 0   1   2   3   4  5  
            # 11 13 17 19 23 29
#                       -- --
# 요소의 갯수가 짝수 일때는 가운데 요소가 2개라는 것을 알수 있다. 
# 이 경우 어느쪽을 선택해야 한다고 정해진 것은 없다. 단지 둘 중 하나만
# 선택하면 된다. 따라서 같은 공식으로 계산해 보면
# (0+5) / 2 = 2.5
# 소수를 정수화 하는 방법으로는 1.반올림 2. 올림 3.버림 있다. 
# 하지만 제일 간단한 방법은 3. 버림 이 제일 쉬운 처리이다. 
# 정수형 변수에 넣기만 하면 자동으로 소수 부분은 잘려 나가기 때문에
# 쉽게 정수 부분만 얻을수 있게 된다. 
```
